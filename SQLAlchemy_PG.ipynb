```python
import tweepy
from decouple import config

auth = tweepy.OAuthHandler(config('CONSUMER_KEY'), config('CONSUMER_SECRET'))
api = tweepy.API(auth)

test_tweet = None

LAT = 40.712776
LONG = -74.005974

# May not work for Gujarati cities. Not enough people posting with location info

geocode = str(LAT) + ',' + str(LONG) + ',100km'
print(geocode)

for tweet in tweepy.Cursor(api.search, q='meme', geocode=geocode).items(100):
    json_dict = tweet._json
    try:
        print(json_dict['place']['name'], '\n')
    except TypeError:
        print('No Location\n')
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=db.engine)
session = Session()
```

```python
hashtags = test_tweet['entities']['hashtags']
print(hashtags)
tags = [Hashtag(hashtag=hashtags[i]['text']) for i in range(len(hashtags))]
for tag in tags:
    print(tag.hashtag)
    session.add(tag)
session.commit()
```

```python
place_id = test_tweet['place']['id']
name = test_tweet['place']['name']
country = test_tweet['place']['country']
country_code = test_tweet['place']['country_code']
place = Place(place_id=1, name=name, country=country, country_code=country_code)
# print(place_id)
session.add(place)
session.commit()
```

```python
# print(test_tweet)
tweet_obj = BaseTweet(tweet_id=test_tweet['id'],
                     tweet_id_str=test_tweet['id_str'],
                     source=test_tweet['source'],
                     favorited=test_tweet['favorited'],
                     retweeted=test_tweet['retweeted'],
                     favorite_count=test_tweet['favorite_count'],
                     retweet_count=test_tweet['retweet_count'],
                     result_type=test_tweet['metadata']['result_type'],
                     created_at=datetime.strptime("21/11/06 16:30", "%d/%m/%y %H:%M"),
                     lang='en',
                     possibly_sensitive=False,
                     reply_count=0,
                     place_id=1)
tweet_obj.place = place
tweet_obj.hashtags = tags
# print(tweet_obj.hashtags)
session.add(tweet_obj)
session.commit()
```
